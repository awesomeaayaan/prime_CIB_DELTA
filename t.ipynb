```python
import json
import pandas as pd

file = open(r"D:\BOT_Prime\CIB_DELTA_back\output\data.json")
datas = json.load(file)
df = pd.DataFrame(columns=[
        'Name',
        'DOB', 
        'Gender', 
        'FatherName', 
        'CitizenshipDetails', 
        'PassportDetails', 
        'DrivingLicenseDetails',
        'VoterIDDetails',
        'PANDetails', 
        'CompanyDetails',
        'IndianEmbassyDetails',
        'BlackLists' 
    ])

num = 0
for data in datas["BulkOffLoading"]["Individuals"]["Item"]:
    # for key_value in data:
    # print("--------------------------- REal Data --------------------- \n")
    # print(data)
    dicts = {}
    data_keys = list(data.keys())
    # print(data_keys)
    for data_key in data_keys:
        # print(data_key)
        if isinstance(data[data_key], str):
            dicts[data_key] = data[data_key]
            print(dicts[data_key])
        
        else:
            # maintain Citizenship details and black list detail
            keys = list(data[data_key].keys())
            data_count = int(data[data_key][keys[0]])
            # print('DATA_COUNT',data_count)
            if data_count > 1: #here data with data_count greater than 1 will come in list format
                details = data[data_key][keys[1]]
                # if details contain 
                print('DETAILS',details)
                dicts[data_key] = ['|'.join(entry.values()) for entry in details] 
                print(dicts[data_key])
                
            elif data_count == 0:
                continue
            else:
                details = data[data_key][keys[1]]
                # print('next detail',details)
                dicts[data_key] = '|'.join(details.values())
            # df.loc[len(df)] = dicts
            
    # print(dicts)
    # print(f"--------------- Processed Data number {num} -----------------------" )
    print('Dictionary==========================================================>',dicts)
    num += 1
    if num == 2:
        break
    df.loc[len(df)] = dicts
```

```python
import numpy as np
def drop_dupicates(df):
    both_present = df[(~df['CitizenshipDetails'].isnull()) & (~df['PANDetails'].isnull())]
    # both_present
    duplicated_rows = both_present.copy()
    # duplicated_rows
    both_present_indices = both_present.index 

    df.drop(index=both_present_indices, inplace=True)
    for index, row in both_present.iterrows():
        both_present.at[index, 'CitizenshipDetails'] = np.nan

    for index, row in duplicated_rows.iterrows():
        duplicated_rows.at[index, 'PANDetails'] = np.nan

    df = pd.concat([df, both_present, duplicated_rows], ignore_index=True)
    return df
data_df = drop_dupicates(df)
```
